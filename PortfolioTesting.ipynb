```python
# Rolling pct change with graph
# Rolling mean with a window period(duration)
# 


stocks = yf.download(TechPort, start = "2017-01-01", end = "2021-10-01")
stocks.head()
```

```python
DataTest = RealTime[['ticker', 'date','pe']]
DataTest
```

```python
#Compare the Price/Equity Values
peDF = DataTest.pivot(index='date', columns='ticker')
print(peDF)
```

```python
peDF = pd.DataFrame(peDF)
peDF.tail(30)
```

```python
peDF.plot(figsize = (15, 8), fontsize = 13)
plt.legend(loc = "upper left",fontsize = 13)
plt.show
```

```python
peDF = peDF.resample("BM").last().pct_change(periods = 1).mul(100)
peDF.tail()
```

```python
peDF.plot(figsize = (15, 8), fontsize = 13, subplots=True)
plt.legend(loc = "upper left",fontsize = 13)
plt.show
```
